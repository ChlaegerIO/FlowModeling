```python
import torch
import torch.nn as nn
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import os
from os import listdir
import cv2
import random
import time
import math
from scipy.signal import savgol_filter
import sys
sys.path.append("src")
import sindy_utils as sindy
from datetime import datetime
print(datetime.now().strftime('%Y-%m-%d %H:%M:%S'))

# autoencoder architecture
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__() 
        self.encode = nn.Sequential(
            # encoder: N, 3, 404, 720
            nn.Conv2d(3, 16, 2), # N, 16, 403, 719
            nn.ReLU(),
            nn.Conv2d(16, 32, 2), # N, 32, 402, 718
            nn.ReLU(),
            nn.MaxPool2d((2,3), stride=(2,3)), # N, 32, 201, 239              -- pool --
            nn.Conv2d(32, 64, 4), # N, 64, 198, 236
            nn.ReLU(),
            nn.Conv2d(64, 96, 4), # N, 96, 195, 233
            nn.ReLU(),
            nn.MaxPool2d(2, stride=2), # N, 96, 97, 116                       -- pool --
            nn.Conv2d(96, 128, 5), # N, 128, 93, 112
            nn.ReLU(),
            nn.Conv2d(128, 150, 5, stride=2, padding=1), # N, 150, 46, 55
            nn.ReLU(),
            nn.MaxPool2d(2,stride=2), # N, 150, 23, 27                        -- pool --
            nn.Conv2d(150, 200, 9, stride=2), # N, 200, 8, 10
            nn.ReLU()
        )
        
        self.fc1 = nn.Linear(200*8*10,gl_zDim)
        # Note: nn.MaxPool2d -> use nn.MaxUnpool2d, or use different kernelsize, stride etc to compensate...
        # Input [-1, +1] -> use nn.Tanh    
        
        # note: encoder and decoder are not symmetric
        self.decode = nn.Sequential(
            nn.ConvTranspose2d(200, 150, 4), # N, 150, 11, 13
            nn.ReLU(),
            nn.ConvTranspose2d(150, 128, 5, stride=(2,3), padding=(2,2), output_padding=(0,2)), # N, 128, 21, 39
            nn.ReLU(),
            nn.ConvTranspose2d(128, 96, 4, stride=2, padding=(1,0)), # N, 96, 42, 80
            nn.ReLU(),
            nn.ConvTranspose2d(96, 64, 8), # N, 64, 49, 87
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 8, stride=2, padding=(2,1), output_padding=(0,1)), # N, 32, 100, 179
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 5, stride=2, padding=1), # N, 16, 201, 359
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 5, stride=2, padding=1, output_padding=(1,1)), # N, 3, 404, 720
            nn.ReLU()
        )   
        
        self.fc2 = nn.Linear(gl_zDim, 200*8*10)

    def forward(self, x, z, mode):
        '''
        x: input for encoder
        z: input for decoder
        mode: 
            'train' -> use encoded for decoder
            'test'  -> feed z in an get decoded
        
        '''
        if mode == 'train':
            encoded = self.encode(x)
            encoded = encoded.view(-1,200*8*10)
            encoded = self.fc1(encoded)

            decoded = self.fc2(encoded)
            decoded = decoded.view(-1,200,8,10)
            decoded = self.decode(decoded)
        else:
            encoded = torch.zeros(1)

            decoded = self.fc2(z)
            decoded = decoded.view(-1,200,8,10)
            decoded = self.decode(decoded)
        
        return encoded, decoded

    
def calculateSindy(z, Xi, poly_order, include_sine_param):
    z_new = z.detach().numpy()
    
    theta = torch.from_numpy(sindy.sindy_library(z_new, poly_order, include_sine=include_sine_param))
    
    dz_prediction = torch.matmul(theta, Xi).float()
    
    return dz_prediction

# use more than 16 frames to get Xi
def constructXi(data, zDim):
    '''
    input: data as a list with shape [len batch_size RGB hight width]
    return: Xi
    
    '''
    # processs the data
    z_tensor = torch.empty((0, zDim))
    data_len = len(data)
    for i in range(data_len):
        if i == 20:
            break
        z_tensor_tmp, _ = autoencoder(train_data[i], 0, mode='train')
        z_tensor = torch.cat((z_tensor, z_tensor_tmp), 0)
        if i % 5 == 0:
            print(i, z_tensor.shape)
        del z_tensor_tmp

    print(z_tensor.shape)
    
    dz_tensor = z_tensor[2:data_len]
    z_tensor = z_tensor[1:data_len-1]
    
    # calculate sindy and Xi for the data
    z = z_tensor.cpu().detach().numpy()
    dz = dz_tensor.cpu().detach().numpy()

    Theta = torch.from_numpy(sindy.sindy_library(z, gl_poly_order, include_sine=gl_include_sine_param))
    Xi = torch.from_numpy(sindy.sindy_fit(Theta, dz, gl_threshold_sindy))
    
    return Xi

def matrixToNorm(x, offset=0, factor=0.95):
    x = (x - x.min() + offset) / x.max() * factor
    return x


def calculate_psnr(img1, img2, border=0):
    '''
    img1 and img2 have range [0, 255]
    '''
    if not img1.shape == img2.shape:
        raise ValueError('Input images must have the same dimensions.')
    h, w = img1.shape[:2]
    img1 = img1[border:h-border, border:w-border]
    img2 = img2[border:h-border, border:w-border]

    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    mse = np.mean((img1 - img2)**2)
    if mse == 0:
        return float('inf')
    return 20 * math.log10(255.0 / math.sqrt(mse))
```

### calculating the PSNR

```python
# note the method is difficult as the fps of the prediction video is not the same as the original video!
whichData = 'train_data'
if whichData == 'train_data':
    data = train_data
elif whichData == 'validation_data':
    data = validation_data
elif whichData == 'test_data':
    data = test_data
    

# load the predicted video also in batch form to be consistent
fileName_predVideo = path_figures + 'prediction_smoothed_' + str(whichData) + '_' + str(gl_epoch) + 'epoch'+gl_addedName+'_Trim.mp4'
print(fileName_predVideo)
#fileName_predVideo = path_figures + 'prediction_smoothed_train_data_5000epoch_batch_Trim.mp4'

if os.path.isfile(fileName_predVideo):
    predictionVideo_tmp = []
    predictionVideo = []
    transform = transforms.ToTensor()

    # start reading the video
    vidcap = cv2.VideoCapture(fileName_predVideo)
    success,imgR = vidcap.read()
    while success:
        imgR = cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB)
        imgR_tensor = transform(imgR)
        predictionVideo_tmp.append(imgR_tensor)
        success,imgR = vidcap.read()
        # make a batch
        if len(predictionVideo_tmp) >= gl_batchSize:
            predictionVideo.append(torch.stack(predictionVideo_tmp))
            del predictionVideo_tmp
            predictionVideo_tmp = []
    print('read prediction video done! ', len(predictionVideo), len(predictionVideo[0]), len(predictionVideo[0][0]), len(predictionVideo[0][0][0]), len(predictionVideo[0][0][0][0]))
    vidcap.release()
else:
    print('video was not read!')

# compare voth videos and trim them to the same length
batchesToDiscard = len(predictionVideo) - len(data)
print('batches to discard ', batchesToDiscard)
predictionVideo_final_tmp = []
predictionVideo_final = []
if batchesToDiscard > 0:
    framesToDiscard = batchesToDiscard * gl_batchSize
    modValue = int(len(predictionVideo)*gl_batchSize / framesToDiscard)
    boolSecondRound = False
    if modValue == 1:
        modValue = 2
        boolSecondRound = True
    print('modulo value ', modValue)
    for f in range(len(predictionVideo)):
        for bs in range(gl_batchSize):
            # dont take this frames: frameNumer % frame to discard
            if (f*gl_batchSize + bs) % modValue != 0:
                predictionVideo_final_tmp.append(predictionVideo[f][bs])
            
            # make a batch
            if len(predictionVideo_final_tmp) >= gl_batchSize:
                predictionVideo_final.append(torch.stack(predictionVideo_final_tmp))
                del predictionVideo_final_tmp
                predictionVideo_final_tmp = []
                
    # second round to discard frames, because mod was < 2
    if boolSecondRound == True:
        batchesToDiscard = len(predictionVideo_final) - len(data)
        print('batches to discard round 2:', batchesToDiscard)
        predictionVideo_final2_tmp = []
        framesToDiscard = batchesToDiscard * gl_batchSize
        modValue = int(len(predictionVideo)*gl_batchSize / framesToDiscard)
        print('modulo value ', modValue)
        for f in range(len(predictionVideo_final)):
            for bs in range(gl_batchSize):
                # dont take this frames: frameNumer % frame to discard
                if (f*gl_batchSize + bs) % modValue != 0:
                    predictionVideo_final_tmp.append(predictionVideo_final[f][bs])
            
                # make a batch
                if len(predictionVideo_final_tmp) >= gl_batchSize:
                    predictionVideo_final2_tmp.append(torch.stack(predictionVideo_final_tmp))
                    del predictionVideo_final_tmp
                    predictionVideo_final_tmp = []
                    
        predictionVideo_final = predictionVideo_final2_tmp
        del predictionVideo_final2_tmp
    
    del predictionVideo
else:
    predictionVideo_final = predictionVideo
print('prediction fps addaption done! ', len(predictionVideo_final), len(predictionVideo_final[0]), len(predictionVideo_final[0][0]), len(predictionVideo_final[0][0][0]), len(predictionVideo_final[0][0][0][0]))
    
    
# compare frame by frame
psnr_perFrame = []
# take smaler length of both videos
maxLength = 0
if len(data) <= len(predictionVideo_final):
    maxLength = len(data)
else:
    maxLength = len(predictionVideo_final)
print('length of graph')
    
# loop over all frames
for f in range(maxLength):
    for bs in range(gl_batchSize):
        psnr = calculate_psnr(data[f][bs].permute(1,2,0).detach().numpy(), predictionVideo_final[f][bs].permute(1,2,0).detach().numpy())
        psnr_perFrame.append(psnr)
        

# plot graph with psnr per frame and calculate mean
plt.figure()
x = np.linspace(0, maxLength*gl_batchSize, maxLength*gl_batchSize)
plt.plot(x, psnr_perFrame, label='psnr per frame')
avg_psnr = sum(psnr_perFrame) / len(psnr_perFrame)
plt.axhline(y=avg_psnr, color='r', label='average psnr')
plt.legend()
plt.savefig(path_figures+'psnr_'+str(whichData)+'_'+str(gl_epoch)+'epoch'+gl_addedName+'.png', transparent=True, bbox_inches="tight")
plt.show()

print('mean psnr is: ', avg_psnr)
```

```python
whichData = 'train_data'
if whichData == 'train_data':
    data = train_data
elif whichData == 'validation_data':
    data = validation_data
elif whichData == 'test_data':
    data = test_data
    
ChooseLength = len(data)                   # len(data)
ChoosePredictionLength = len(data)      # len(data)
batch_size = 16

# compute true z
z_true = np.empty((ChooseLength*batch_size, gl_zDim))
for index in range(0, ChooseLength):
    if index >= len(data):
        break
    z_tensor, recon_tensor = autoencoder(data[index], 0, mode='train')
    z_true[index*batch_size:(index+1)*batch_size][:] = z_tensor.cpu().detach().numpy()
    if index % 5 == 0: print(index)

print('z_true dimension', z_true.shape)
    
# print graph(s) for z_true
for graph in range(z_true.shape[1]):
    plt.figure(graph)
    x = np.linspace(0, z_true.shape[0], z_true.shape[0])
    y = z_true[:,graph]
    plt.plot(x, y, color='black', label='z true')

# compute sindy z predition based on first one
z_tensor, recon_tensor = autoencoder(data[0], 0, mode='train')
z = z_tensor.cpu().detach().numpy()
z_sindy = np.empty((ChoosePredictionLength*batch_size, gl_zDim))
z_sindyUnsmooth = np.empty((ChoosePredictionLength*batch_size, gl_zDim))
z_sindy[0:16] = z
for index_s in range(0, ChoosePredictionLength-1):
    Theta = torch.from_numpy(sindy.sindy_library(z, gl_poly_order, include_sine=gl_include_sine_param))
    dz_predict = torch.matmul(Theta, Xi).float()
    z_sindy[(index_s+1)*batch_size:(index_s+2)*batch_size] = dz_predict.cpu().detach().numpy()
    z = dz_predict
        
# print graph(s) for z_sindy
for graph in range(z_sindy.shape[1]):
    plt.figure(graph)
    x = np.linspace(0, z_sindy.shape[0], z_sindy.shape[0])
    y = z_sindy[:,graph]
    z_sindyUnsmooth[:,graph] = z_sindy[:,graph]              # save unsmoothed signal for later use
    plt.plot(x, y, label='z prediction')
    # smooth z coordinate plot
    z_sindy[:,graph] = savgol_filter(z_sindy[:,graph], 61, 3)
    y = z_sindy[:,graph]
    plt.plot(x, y, color='red', label='z smoothed')
    plt.legend()
    plt.savefig(path_figures+'graph_z'+str(graph)+'_'+str(whichData)+'_length'+str(ChoosePredictionLength)+'_'+str(gl_epoch)+'epoch'+gl_addedName+'.png', transparent=True, bbox_inches="tight")

plt.show()

del z_true
```
